## Approximation 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

### REGRESSION

```python
reg = np.polyfit(x, f(x), deg=1)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
reg = np.polyfit(x, f(x), 7)
ry = np.polyval(reg, x)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

### A brief check reveals that the result is not perfect:

```python
np.allclose(f(x), ry)
```

### the mean squared error (MSE) is not too large—at least, over this narrow range of x values:

```python
np.sum((f(x) - ry) ** 2) / len(x)
```

### Individual basis functions

```python
matrix = np.zeros((3 + 1, len(x)))
matrix[3, :] = x ** 3
matrix[2, :] = x ** 2
matrix[1, :] = x
matrix[0, :] = 1
```

```python
#### The sublibrary numpy.linalg provides the function lstsq to solve least-squares optimization problems
```

```python
reg = np.linalg.lstsq(matrix.T, f(x))[0]
```
